# Rankeamento de Ações com base em indicadores

## EV-EBIT

```python
# Abre o dataset
ev_ebit = pd.read_excel('../Bruto/EVEBIT.xlsx', Header = None)
ev_ebit
```

```python
# Removendo informações indesejadas
ev_ebit = ev_ebit.drop([0])
ev_ebit = ev_ebit.drop([211,212,213,214,215,216,217,218,219,220])

# Trocando NaN por 0

ev_ebit = ev_ebit.fillna(0)
```

```python
# Conferindo
ev_ebit
```

## A intenção é criar um dataframe, onde as datas são linhas e as colunas estão ordenadas pelos melhores valores

Para isso, devemos:
- analisar cada data (linha)
- remover os outliers
- adicionar os 10 primeiros a um dataframe "Ranking"

```python
ranking = pd.DataFrame()
for i in range (2,len(ev_ebit)): #len(ev_ebit)-10
    acoes_data = ev_ebit.loc[i]
    data = acoes_data.loc['Data']
    data = str(data)
    acoes_data = acoes_data.drop(labels='Data')
    acoes_data = acoes_data.sort_values(ascending=True)
    acoes_data = acoes_data[acoes_data < 3000]
    acoes_data = acoes_data[acoes_data > 0]
    acoes = acoes_data.index
    acoes = acoes.tolist()
    adicional = pd.DataFrame({data:acoes})
    ranking = pd.concat([ranking, adicional], ignore_index=False, axis=1) 
    
ranking
```
